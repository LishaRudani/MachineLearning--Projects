```python
x = redmi_smartphone.drop(['title', 'ram', 'brand', 'url', 'product_id', 'listing_id','highlights',
'availability', 'selling_price', 'currency'], axis = 1)
y = redmi_smartphone['selling_price']
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
model_india=LinearRegression()
model_india.fit(x,y)
```

```python
print("Training Accuracy :", model_india.score(X_train, y_train))
print("Testing Accuracy :", model_india.score(X_test, y_test))
```

```python
print("Coefficient: ",model_india.coef_)
print("intercept: ",model_india.intercept_)
pre = model_india.predict(x)
```

```python
from sklearn.metrics import mean_squared_error
print("By function: ",mean_squared_error(y,model_india.predict(x)))
```
